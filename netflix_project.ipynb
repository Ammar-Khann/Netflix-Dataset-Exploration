```python
pdf['date_added'] = pdf['date_added'].str.strip()
pdf['date_added'] = pd.to_datetime(pdf['date_added'])
```

```python
# remove the international Tv show and movies
def clean_listed_in(listed_in):
    if pd.isna(listed_in):
        return listed_in
    clean = [c.strip() for c in listed_in.split(',')]
    clean = [c for c in clean if c != "International TV Shows"]
    return ','.join(clean) if clean else "unknown"

pdf['listed_in'] = pdf['listed_in'].apply(clean_listed_in)

def clean_inter_movies(listed_in):
    if pd.isna(listed_in):
        return listed_in
    clean = [c.strip() for c in listed_in.split(',')]
    clean = [c for c in clean if c != "International Movies"]
    return ','.join(clean) if clean else "unknown"

pdf['listed_in'] = pdf['listed_in'].apply(clean_inter_movies)
```

```python
def clean_british(listed_in):
    if pd.isna(listed_in):
        return listed_in
    clean = [c.strip() for c in listed_in.split(',')]
    clean = [c for c in clean if c != "British TV Shows"]
    return ','.join(clean) if clean else "unknown"

pdf['listed_in'] = pdf['listed_in'].apply(clean_british)

def clean_spanish(listed_in):
    if pd.isna(listed_in):
        return listed_in
    clean = [c.strip() for c in listed_in.split(',')]
    clean = [c for c in clean if c != "Spanish-Language TV Shows"]
    return ','.join(clean) if clean else "unknown"

pdf['listed_in'] = pdf['listed_in'].apply(clean_spanish)

```

```python
def clean_lgbt(listed_in):
    if pd.isna(listed_in):
        return listed_in
    clean = [c.strip() for c in listed_in.split(',')]
    clean = [c for c in clean if c != "Untold: Deal With the Devil"]
    return ','.join(clean) if clean else "unknown"

pdf['listed_in'] = pdf['listed_in'].apply(clean_lgbt)
```

```python
pdf = pdf.drop('cast', axis = 1)
pdf = pdf.drop('description', axis = 1)
```

```python
# creating  the genre column
def genre(listed_in):
    if pd.isna(listed_in):
        return "Unknown"
    elif "Documentary" in listed_in or "Documentaries" in listed_in or "Docuseries" in listed_in:
        return "Documentary"
    elif "Mysteries" in listed_in or "Thrillers" in listed_in:
        return "Thriller"
    elif "Crime" in listed_in or "Crime TV Shows" in listed_in:
        return "Crime"
    elif "TV Comedies" in listed_in or "Stand-Up Comedy & Talk Shows" in listed_in or "Comedies" in listed_in:
        return "Comedy"
    elif "Action & Adventure" in listed_in:
        return "Action & Adventure"
    elif "Sci-Fi & Fantasy" in listed_in or "TV Sci-Fi & Fantasy" in listed_in:
        return "Sci-Fi & Fantasy"
    elif "Romantic Movies" in listed_in or "Romantic TV Shows" in listed_in or "TV Dramas" in listed_in or "Dramas" in listed_in:
        return "Drama"
    elif "Children & Family Movies" in listed_in or "Kids' TV	":
        return "Kids' show"
    else:
        return "Other"


pdf['genre'] = pdf['listed_in'].apply(genre)
```

```python
pdf['listed_in'] = pdf['listed_in'].str.replace('Docuseries', 'Documentaries')
```
